```python
from transformers import BertTokenizer,BertModel
"""
匯入BertTokenizer和BertModel，
BertTokenizer將文本轉換為模型可理解的編碼，
BertModel 是 BERT 預訓練模型。
"""
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
#加載了預訓練的BERT分詞器，將輸入文本轉換為BERT所需的編碼格式。
model = BertModel.from_pretrained("bert-base-uncased")
#加載了預訓練的BERT模型，用於處理分詞器編碼後的輸入。
text="my bank account"
encoded_input = tokenizer(text, max_length=100,
         add_special_tokens=True, truncation=True,
         padding=True, return_tensors="pt")
"""
將文本轉換為BERT模型的輸入格式。
add_special_tokens=True表示在文本的開始和結尾添加特殊tokens。
truncation=True表示如果文本長度超過最大長度則進行截斷。
padding=True表示如果文本長度小於最大長度則進行填充。
"""
output = model(**encoded_input)
#將編碼後的輸入傳遞給BERT模型，並獲取模型的輸出。
print(output[0].shape)
#打印模型輸出張量的形狀。
print(output)
#打印模型的完整輸出。
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
"""
匯入GPT2Tokenizer和GPT2LMHeadModel，並匯入PyTorch的torch庫。
GPT2Tokenizer用於將文本轉換為模型可理解的編碼，
GPT2LMHeadModel是帶有語言模型頭部的GPT-2預訓練模型的類。
"""

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
#加載預訓練的GPT-2模型（帶有語言模型頭部的版本）和分詞器

text = "The quick brown fox jumps over the lazy dog"
inputs = tokenizer(text, return_tensors="pt")
#準備了要處理的文本，並將其轉換為GPT-2模型的輸入格式。

with torch.no_grad():
    outputs = model(**inputs)
#使用torch.no_grad()來禁用梯度計算，並將編碼後的輸入傳遞給GPT-2模型，獲取模型的輸出。

logits = outputs.logits
"""
提取模型的logits。
logits是應用softmax之前的輸出，直接對應於詞彙表的維度
"""

print("Logits shape:", logits.shape)
#展示logits的維度
#維度應該是 [批次大小, 序列長度, 詞彙表大小]
```

```python
from transformers import pipeline
#匯入pipeline函式，使用預訓練模型進行各種NLP任務。

unmasker = pipeline("fill-mask", model="bert-base-uncased")
#創建了一個名為unmasker的pipeline，用於填補掩碼（fill-mask）任務。
result = unmasker("This man works as a [MASK].")
#使用unmasker pipeline對給定的句子進行填補掩碼任務。
print([r["token_str"] for r in result])
#打印模型對掩碼標記的預測結果。

result = unmasker("This woman works as a [MASK].")
#使用unmasker pipeline對另一個句子進行填補掩碼任務。
print([r["token_str"] for r in result])
#打印模型對第二個句子中掩碼標記的預測結果。
```
