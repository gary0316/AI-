```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
"""
導入了AutoModelForCausalLM、AutoTokenizer和BitsAndBytesConfig。
AutoModelForCausalLM用於加載自回歸語言模型，
AutoTokenizer用於加載分詞器，
BitsAndBytesConfig用於配置量化設置。
"""

from peft import AutoPeftModelForCausalLM
#導入了AutoPeftModelForCausalLM，專門為支持量化的自回歸語言模型而設計的。
import torch
#導入 torch 庫，用於深度學習。

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
"""
load_in_4bit=True：啟用 4-bit 量化。
bnb_4bit_use_double_quant=True：啟用雙重量化。
bnb_4bit_quant_type="nf4"：設定量化類型為 nf4。
bnb_4bit_compute_dtype=torch.bfloat16：設置計算類型為 bfloat16。
"""
peft_model_id = "huchiahsi/viggo-peft-model"
#設置了模型的ID，即"huchiahsi/viggo-peft-model"，用於從模型庫中加載特定的模型。

model = AutoPeftModelForCausalLM.from_pretrained(peft_model_id, device_map="auto", quantization_config=bnb_config)
#使用AutoPeftModelForCausalLM從預訓練模型中加載指定的模型peft_model_id，並配置設備映射為auto和量化配置bnb_config。
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-v0.1")
#使用AutoTokenizer加載名稱為"mistralai/Mistral-7B-v0.1"的預訓練分詞器。
```

```python
eval_prompt = """Given a target sentence construct the underlying meaning representation of the input sentence as a single function with attributes and attribute values.
This function should describe the target string accurately and the function must be one of the following ['inform', 'request', 'give_opinion', 'confirm', 'verify_attribute', 'suggest', 'request_explanation', 'recommend', 'request_attribute'].
The attributes must be one of the following: ['name', 'exp_release_date', 'release_year', 'developer', 'esrb', 'rating', 'genres', 'player_perspective', 'has_multiplayer', 'platforms', 'available_on_steam', 'has_linux_release', 'has_mac_release', 'specifier']

### Target sentence:
這個賽車遊戲叫大賽車，可以多人玩，好玩，我可以Linux上和蘋果電腦上玩。

### Meaning representation:
"""
#定義了一個多行字串變數eval_prompt，用於構建一個目標句子的語義表示

model_input = tokenizer(eval_prompt, return_tensors="pt").to("cuda")
#將評估提示字串eval_prompt進行分詞並轉換為PyTorch張量，然後將其移動到GPU上，準備用於模型推理。
```

```python
model.eval()
#將模型設置為評估模式。
with torch.no_grad():
#創建了一個上下文管理器，在該上下文中，所有的操作都不會計算梯度。
    print(tokenizer.decode(model.generate(**model_input, max_new_tokens=100, pad_token_id=2)[0], skip_special_tokens=True))
    #在模型評估模式下無需計算梯度地生成文本，然後將生成的標記序列解碼並打印出來。
```
